```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
with open('adjacency_matrix.txt') as f:
  lines = f.readlines()

graph = []

for line in lines:
  row = [int(i) for i in line.split()]
  graph.append(row)
```

```python
G = nx.from_numpy_matrix(np.matrix(graph), create_using=nx.DiGraph)
layout = nx.spring_layout(G)
nx.draw(G, layout, with_labels=True)
nx.draw_networkx_edge_labels(G, pos=layout)
plt.show()
```

```python
minCost = sys.maxsize
minPath = []
v = 5
path = []
visited = [False for i in range(5)]
```

```python
def printLine():
    print("-----------------------------------------------------------------")
```

```python
def dfs(graph,source,destination,vis,path,currSum):
    global minPath,minCost
    if(source==destination):
        path.append(destination)
        if(minCost > currSum):
            minCost = currSum
            minPath[:] = path
        print("Path : ", end = " ")
        for i in range(len(path)):
            if i == len(path)-1:
                print(path[i])
            else:
                print(path[i],end="-->")
        print("Total path sum : ",currSum)
        path.pop()
        printLine()
        return
    
    path.append(source)
    vis[source]=True
    for n in range(v):
        if(vis[n]==False):
            dfs(graph,n,destination,vis,path,currSum+graph[source][n])
    
    vis[source]=False
    path.pop()
```
